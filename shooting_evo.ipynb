### The point of the analysis is to show the evolution of NBA shooting throughout the last ~25 years, so of course I will mainly need the data grouped by season.

```python
by_season=d1.groupby('season')
```

### Let's start by looking at the overall shooting efficiency. In general it has been increasing with an over 2 percentage point drop around 2012 (perhaps the effect of shortened lockout season?). Current shooting efficiency is by far the highest in the last quarter-century with ~46.5%.

```python
by_season['fg_percent'].mean().plot(lw=3,title='Overall shooting efficiency',colormap='Dark2',xlabel='Season',\
                                    ylabel='Field Goal %',grid=True);
```

### When we look at the number of players shooting over 50% from the field, the graph turns out to be quite similar to the previous one, with the number of >50% shooters reaching 100 this season.

```python
d1.query("fg_percent>=0.5 & games>=41").groupby('season').count()['player'].plot(lw=3,title='Number of 50%+ FG players',colormap='Dark2',xlabel='Season',\
                                    ylabel='Number of players',grid=True);
```

### One could think that's because players have been shooting from closer to the basket, right? It's easier to score standing closer, isn't it? Well, let's see....

```python
by_season['avg_dist_fga'].mean().plot(lw=3,title='Average shot distance',colormap='Dark2',xlabel='Season',\
                                      ylabel='Distance in feet',grid=True);
```

### The evolution of shooting assumes shooting from further away, almost 3 feet further compared to ~2002/2003. When we take a look at the number of 2-point shots attempted, the drop has been quite shocking and very constant - basically 25 pp since the end of the Jordan era.

```python
by_season['percent_fga_from_x2p_range'].mean().plot(lw=3,title='% of 2-pointers taken',colormap='Dark2',xlabel='Season',\
                                    ylabel='% of total shots taken',grid=True);
```

### Shooting from the paint, however, has remained stable, at a little less than 50% of all shots taken.

```python
ax=by_season[['percent_fga_from_x0_3_range','percent_fga_from_x3_10_range']].mean().plot(kind='area',title='Shooting from the paint',\
                                                                                      colormap='Dark2',xlabel='Season',\
                                                                                      ylabel='% of total shots taken',\
                                                                                         grid=True)
h,l=ax.get_legend_handles_labels()
ax.legend(h[0:2],['%FGA from 0-3ft','%FGA from 3-10ft range']);
```

### And here comes the midrange, the archenemy of modern basketball. Since 2004 it has been going down constantly, from 36-37% of total shots taken to slightly less than 15%. It is the least efficient type of shot and teams know it.

```python
ax=by_season[['percent_fga_from_x10_16_range','percent_fga_from_x16_3p_range']].mean().plot(kind='area',title='Shooting from mid range',\
                                                                                      colormap='Dark2',xlabel='Season',\
                                                                                      ylabel='% of total shots taken',\
                                                                                     grid=True)
h,l=ax.get_legend_handles_labels()
ax.legend(h[0:2],['%FGA from 10-16ft','%FGA from 16ft-3pt line'],loc='lower left');
```

### And why is it so inefficient? Let's look:

```python
by_season[['fg_percent_from_x10_16_range','fg_percent_from_x16_3p_range']].mean().plot(lw=3,title='Midrange shooting efficiency',colormap='Dark2',xlabel='Season',\
                                      ylabel='FG%',grid=True)
plt.legend(['10-16ft','16ft-3pt line'],loc='lower right');
```

### Basically an under 40% shot worth only 2 points. But we know what an efficient shot is, right? It is the dunk:

```python
by_season['percent_dunks_of_fga'].mean().plot(lw=3,title='% of dunks',colormap='Dark2',xlabel='Season',\
                                    ylabel='% of total shots taken',grid=True);
```

### The dunks have been on the rise, as have the 3-pointers:

```python
by_season['percent_fga_from_x3p_range'].mean().plot(lw=3,title='% of 3-pointers taken',colormap='Dark2',xlabel='Season',\
                                    ylabel='% of total shots taken',grid=True);
```

### 40% of total shots taken is an insane number. The rise has been extremely fast and we all owe it to the one and only, Stephen Curry. Let's take a look at the corner threes now:

```python
by_season['percent_corner_3s_of_3pa'].mean().plot(lw=3,title='% of corner 3s taken',colormap='Dark2',xlabel='Season',\
                                    ylabel='% of total 3s taken',grid=True);
```

### Corner threes were recognized as en efficient shot relatively early. There's no rocket science to that: it's the closest to the basket type of three. More than a quarter of all threes have come from the corners almost every year since around 2005.

### More threes attempted should mean a decline in efficiency, shouldn't it? Well, no. Players take more threes because they are getting better and better at them:

```python
axes=by_season[['percent_fga_from_x3p_range','fg_percent_from_x3p_range']].mean().plot(lw=3,subplots=True,\
                                                                                       title='% of 3-pointers taken vs 3FG%',\
                                                                                       colormap='Dark2',\
                                                                                       xlabel='Season',\
                                                                                       grid=True,\
                                                                                      legend=False)
axes[0].set_ylabel('% total shots taken')
axes[1].set_ylabel('3P FG%');
```

### When it comes to team play, the number of shots assisted has been on a delicate rise for the 3-pointers, from around 83% at the and of the 20th century to around 88% now. As far as the twos are concerned, the difference is not big, from ~56% in 1997 to 55% now:

```python
by_season[['percent_assisted_x2p_fg','percent_assisted_x3p_fg']].mean().plot(lw=3,title='% of FG assisted',colormap='Dark2',xlabel='Season',\
                                      ylabel='% FG assisted',grid=True)
plt.legend(['2-pointers','3-pointers'],loc='center');
```

```python
field_goal_pcts=d1[['player','season','fg_percent_from_x2p_range','fg_percent_from_x0_3_range','fg_percent_from_x3_10_range','fg_percent_from_x10_16_range','fg_percent_from_x16_3p_range','fg_percent_from_x3p_range']]
```

```python
field_goal_pcts.head()
```

```python
field_goal_pcts_season=field_goal_pcts.groupby('season')
```

### To sum it up, we can take a look at a graph showing that shooting efficiency has risen no matter the type of shot. The players are simply getting better.

```python
field_goal_pcts_season.mean().plot(figsize=(20,15),lw=5,grid=True,colormap='Dark2')
plt.title('General shooting efficiency',fontsize=35)
plt.xlabel('Season',fontsize=20)
plt.ylabel('FG%',fontsize=20)
plt.tick_params(axis='both', labelsize=20)
plt.legend(labels=['FG% 2-pointers','FG% from 0-3ft','FG% from 3-10ft','FG% from 10-16ft','FG% from 16ft-3pt line','FG% 3-pointers'],
           title='Shot type',title_fontsize=25,fontsize=20,bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0,\
          frameon=False);
```

### As a conclusion we can definitely say that teams have practically eliminated the midrange game at the cost of the 3-point shot. Shooting from close to the basket has always been known to be effective (again, no rocket science) but the fact that the 3-point shot is literally worth three points has drastically changed the game and forced players to improve their skillsets to adjust to modern offenses.
